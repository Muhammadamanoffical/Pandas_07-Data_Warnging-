## horizontal for concatenation  side by side :-
    -increase in columns so x=1
    vertical for concatenation buttom side:-
        -increase in row so y=0

```python
# by default axis is zero it means increasing in rows
pd.concat([df_02,df_01] , ignore_index=True)
```

```python
pd.concat([df_02,df_01], axis=1)
```

# JOINS:-

```python
#  records have same column are called overlapping columns
```

## Inner join

```python
pd.merge(df_03,df_01)
#inner, outer, and right left
# inner means interaction  (by default)
```

```python
df_01.columns=['sales_man_id', 'name', 'sales_man_city', 'product_id',
       'manager']
```

```python
df_01
```

```python
pd.merge(df_03,df_01)
```

```python
pd.merge(df_03,df_01 ,left_on="sales_man_name" , right_on="name").drop("name" ,axis=1)
```

```python

    pd.merge(df_03,df_01 ,left_on="sales_man_name" , right_on="name" ,how="inner").drop("name" ,axis=1)
```

## outer join

```python
pd.merge(df_03,df_01 ,left_on="sales_man_name" , right_on="name" ,how="outer").drop("name" ,axis=1)
```

## left join  

```python
pd.merge(df_03 ,df_01 ,left_on="sales_man_name" , right_on="name" , how="left").drop("name" ,axis=1)
```

```python
df_03
```

```python
pd.merge(df_03,df_01, left_on="sales_man_name" ,right_on="name" ,how="right")
```

### merge by three column

```python
pd.merge(df_01,df_02 , left_on="name" ,right_on="sales_man_name" ,how="outer").merge(df_03)
```

```python
#suffixes
pd.merge(df_02,df_01 , left_on="manager" ,right_on="manager" ,suffixes=("_left" ,"_right"))
```
